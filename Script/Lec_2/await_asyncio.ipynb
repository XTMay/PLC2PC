### ğŸ“Œ ä¸‰ã€ä»€ä¹ˆæ˜¯ awaitï¼Ÿ

await ç”¨äºåœ¨ åç¨‹å†…éƒ¨è°ƒç”¨å¦ä¸€ä¸ªåç¨‹æˆ–å¼‚æ­¥æ“ä½œï¼Œå®ƒçš„ä½œç”¨æ˜¯ â€œæš‚åœå½“å‰åç¨‹æ‰§è¡Œï¼Œè®©å‡ºæ§åˆ¶æƒç»™äº‹ä»¶å¾ªç¯â€ï¼Œç›´åˆ°è¢«ç­‰å¾…çš„åç¨‹å®Œæˆåç»§ç»­æ‰§è¡Œã€‚

```python
import asyncio

# async å…³é”®å­—ç”¨äºå£°æ˜ä¸€ä¸ªå¼‚æ­¥å‡½æ•°ï¼ˆåç¨‹ï¼‰ã€‚
# å®ƒçš„ä½œç”¨æ˜¯å‘Šè¯‰è§£é‡Šå™¨ï¼šè¿™ä¸ªå‡½æ•°ä¸æ˜¯æ™®é€šçš„åŒæ­¥å‡½æ•°ï¼Œè€Œæ˜¯ä¸€ä¸ªå¯ä»¥è¢«æŒ‚èµ·ã€ç­‰å¾…å’Œå¹¶å‘æ‰§è¡Œçš„åç¨‹å‡½æ•°ã€‚
async def boil_water():
    print("å¼€å§‹çƒ§æ°´...")
    await asyncio.sleep(3)  # æ¨¡æ‹Ÿçƒ§æ°´éœ€è¦3ç§’
    print("æ°´çƒ§å¥½äº†ï¼")
    return "çƒ­æ°´"

async def make_tea():
    print("å‡†å¤‡æ³¡èŒ¶...")
    water = await boil_water()  # await æš‚åœè¿™é‡Œï¼Œç­‰ boil_water å®Œæˆ
    print(f"ä½¿ç”¨ {water} æ³¡èŒ¶")
    return "èŒ¶æ³¡å¥½äº†ï¼"

# asyncio.run(make_tea())
# âœ… ç›´æ¥ awaitï¼ˆåªé€‚ç”¨äº Jupyter æˆ–å·²è¿è¡Œäº‹ä»¶å¾ªç¯çš„ç¯å¢ƒï¼‰
await make_tea()
```

```python
import asyncio
# import time

# time.sleep(2)

# å¼‚æ­¥å‡½æ•°å®šä¹‰
async def say_hello(name, delay):
    print(f"â³ æ­£åœ¨ç­‰å¾… {delay} ç§’åå‘ {name} æ‰“æ‹›å‘¼...")
    await asyncio.sleep(delay)
    print(f"ğŸ‘‹ ä½ å¥½ï¼Œ{name}ï¼")

# ä¸»å‡½æ•°ï¼šå¹¶å‘æ‰§è¡Œå¤šä¸ªä»»åŠ¡
async def main():
    # ä½¿ç”¨ asyncio.create_task å¯åŠ¨å¤šä¸ªå¼‚æ­¥ä»»åŠ¡
    task1 = asyncio.create_task(say_hello("Alice", 2))
    task2 = asyncio.create_task(say_hello("Bob", 1))
    task3 = asyncio.create_task(say_hello("Charlie", 3))

    print("ğŸŒŸ æ‰€æœ‰æ‰“æ‹›å‘¼çš„ä»»åŠ¡å·²å¯åŠ¨ï¼ˆéé˜»å¡ï¼‰")

    # ç­‰å¾…æ‰€æœ‰ä»»åŠ¡æ‰§è¡Œå®Œæˆ
    await task1
    print("Alice åšå®Œï¼ï¼")
    await task2
    print("Bob åšå®Œï¼ï¼")
    await task3
    print("Charlie åšå®Œï¼ï¼")

# åœ¨è„šæœ¬æˆ–ç»ˆç«¯è¿è¡Œï¼š
# asyncio.run(main()) â† ä»…é™é Jupyter ç¯å¢ƒ
# åœ¨ Jupyter / Colab / Notebook ä¸­è¿è¡Œï¼š
await main()
```

2. å¹¶å‘è¿è¡Œå¤šä¸ªåç¨‹ï¼ˆéé˜»å¡å¹¶è¡Œï¼‰

```python
import asyncio

async def turn_on_light(name, delay):
    print(f"ğŸ’¡ {name}ï¼šç­‰å¾… {delay} ç§’åå¼€ç¯")
    await asyncio.sleep(delay)
    print(f"âœ… {name} å·²å¼€å¯ï¼")

async def main():
    # åˆ›å»ºå¤šä¸ªå¼‚æ­¥ä»»åŠ¡
    tasks = [
        asyncio.create_task(turn_on_light("çº¢ç¯", 2)),
        asyncio.create_task(turn_on_light("ç»¿ç¯", 3)),
        asyncio.create_task(turn_on_light("è“ç¯", 1))
    ]

    # å¹¶å‘æ‰§è¡Œæ‰€æœ‰ä»»åŠ¡ï¼ˆéé˜»å¡ç­‰å¾…ï¼‰
    await asyncio.gather(*tasks)

# ğŸ“Œ Jupyter ä¸­ç›´æ¥ä½¿ç”¨ awaitï¼Œä¸ç”¨ asyncio.run()
await main()
```
